```python
from bs4 import BeautifulSoup
import requests
```

```python
website_url = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup= BeautifulSoup(website_url,'lxml')
print(soup.prettify())
```

```python
My_table = soup.find('tbody')
tables = My_table.findAll('p')
tables
```

```python
postcodes = My_table.findAll('b')
postcodes
```

```python
for postcode in postcodes:
    bs = postcode.find_all('b')
```

```python
# Search through the tables for the one with the headings we want.
for table in tables:
    ths = table.find_all('th')
    headings = [th.text.strip() for th in ths]
    if headings[:3] == ['Postcode', 'Borough', 'Neighbourhood']:
        break
        
```

```python
# Extract the columns we want and write to a comma-separated file.
##with open(r'/Users/chalbasch/Desktop/data.csv','w+') as fo: ##need Python 3
    for tr in My_table.findAll('tr'):
            tds = tr.findAll('td')
            if not tds:
                continue
            postcode, borough, neighborhood = [td.text.strip() for td in tds [:3]]
   
            print(', '.join([postcode, borough, neighborhood]))#, file=fo) ##need Python 3
    
```

```python
#Removing unassigned Postcodes
df = df[df.Borough != ' Not assigned']
df.head()
```

```python
# combine multiple neighborhoods with the same post code
df2 = pd.DataFrame({'Postcode':df.Postcode.unique()})
df2['Borough']=pd.DataFrame(list(set(df['Borough'].loc[df['Postcode'] == x['Postcode']])) for i, x in df2.iterrows())
df2['Neighborhood']=pd.Series(list(set(df['Neighborhood'].loc[df['Postcode'] == x['Postcode']])) for i, x in df2.iterrows())
df2['Neighborhood']=df2['Neighborhood'].apply(lambda x: ', '.join(x))
df2.dtypes

df2
```
